```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from joblib import dump
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('training/train/train_data_labeled.csv')
df.loc[df['deforested'] == 0, 'deforested'] = 'Forested'
df.loc[df['deforested'] == 1, 'deforested'] = 'Deforested'
df.head()
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df, hue='deforested')
plt.show()
```

```python
rgb_arr = df.drop(columns=['id', 'deforested']).to_numpy()
labels = df['deforested'].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(rgb_arr, labels, random_state=0)

training_accuracy = []
test_accuracy = []
k_range = range(1,20)
neighbors_settings = k_range

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="Training Accuracy")
plt.plot(neighbors_settings, test_accuracy, label="Testing Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("K Neighbors")
plt.legend()
plt.xticks(np.arange(min(k_range), max(k_range)+1, 1.0))
plt.show()
```

```python
# Save Classifier object
n_neighbors = 7
clf = KNeighborsClassifier(n_neighbors=n_neighbors)
clf.fit(X_train, y_train)
dump(clf, 'knn_classifier.joblib')
```
